```python
def solution(N, stages):
    
    stages_dic = dict()
    participants = len(stages)
    answer = []
    
    for i in range(1, N+1):
        fail = stages.count(i)
        if fail == 0:
            value = 0
        else:
            value = fail / participants
        participants -= fail
        stages_dic[i] = value
    
    answer = sorted(stages_dic.items(), key = lambda x : x[-1], reverse = True)
    return [a[0] for a in answer]
```

```python
solution(5, [2, 1, 2, 6, 2, 4, 3, 3])
```
